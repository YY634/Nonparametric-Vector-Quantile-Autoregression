```python
import mne
import numpy as np
import pandas as pd
from pathlib import Path

input_file = r"C:/Users/Desktop/NVQAR/open_eyes/sub-001_task-photomark_eeg.set"
output_dir  = Path(r"C:/Users/Desktop/NVQAR/test_output")
output_dir.mkdir(parents=True, exist_ok=True)

base = Path(input_file).stem
print(f"Processing: {base}")

raw = mne.io.read_raw_eeglab(input_file, preload=True, verbose=False)
data, times = raw.get_data(return_times=True)
ch_names = raw.ch_names

print(f"  Shape : {data.shape}  (channels × timepoints)")
print(f"  Channels: {ch_names}")

np.save(output_dir / f"{base}-ts.npy", data)

with open(output_dir / f"{base}-channels.txt", "w") as f:
    for i, ch in enumerate(ch_names):
        f.write(f"Row {i+1}: {ch}\n")

# # ── R: .csv with row names = electrodes, columns = timestamps ─────────────────
# df = pd.DataFrame(
#     data,
#     index=ch_names,
#     columns=[f"{t:.4f}s" for t in times]
# )
# df.index.name = "channel"
# df.to_csv(output_dir / f"{base}-ts.csv")

print(f"\nDone! Files saved to: {output_dir}")
print(f"  {base}-ts.npy")
print(f"  {base}-channels.txt")
print(f"  {base}-ts.csv")
```

```python
import numpy as np
import glob
from pathlib import Path

folder = r"C:/Users/Desktop/NVQAR/open_eyes"
npy_files = sorted(glob.glob(folder + "/*.npy"))

ch_names = ['Fp1', 'Fp2', 'F3', 'F4', 'C3', 'C4', 'P3', 'P4', 'O1', 'O2',
            'F7', 'F8', 'T3', 'T4', 'T5', 'T6', 'Fz', 'Cz', 'Pz']

print(f"Found {len(npy_files)} .npy files\n")

for f in npy_files:
    data = np.load(f)
    total_nan = np.isnan(data).sum()
    name = Path(f).name
    
    if total_nan == 0:
        print(f"✓ {name}: all clean")
    else:
        print(f"✗ {name}: {total_nan} NaNs found")
        for i, ch in enumerate(ch_names):
            n = np.isnan(data[i]).sum()
            if n > 0:
                print(f"    → {ch}: {n} NaNs")
```
